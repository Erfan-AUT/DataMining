```python
def draw_scatter(y: str, file_name: str, lim=False):
    fig, ax = plt.subplots()
    colors = {'isolated': 'gray', 'deceased': 'red', 'released': 'green'}
    grouped = df.groupby('state')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='birth_year', y=y, label=key, color=colors[key])
    # if lim:
    #     plt.ylim(-7.5)
    plt.savefig("../plots/" + file_name + ".png", bbox_inches='tight', dpi=300)
```

```python
draw_scatter('confirmed_date', 'scatter_by_date')
draw_scatter('sex', 'scatter_sex', lim=True)
```
